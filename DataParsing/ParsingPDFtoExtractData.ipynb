```python
directory = 'image'

th=0
ls=[]

for filename in os.listdir(directory):
    file_name = os.path.join(directory, filename)

    cfg = config.PipelinesConfig()

    cfg.width_range = (100,500)
    cfg.height_range = (100,800)
    cfg.scaling_factors = [0.7]
    cfg.wh_ratio_range = (1, 3)
    cfg.group_size_range = (2, 100)
    cfg.dilation_iterations = 0
    
    rects, grouping_rects, image, output_image = get_boxes(file_name, cfg=cfg, plot=False)
    
    for rect in rects:
        dic={}
        th+=1
        x=rect[0] 
        y=rect[1]
        w=rect[2]
        h=rect[3]

        image=cv2.imread(file_name)
        crop_img = image[y:y+h, x:x+w]  
        img = Image.fromarray(crop_img)  

        pytesseract.pytesseract.tesseract_cmd ='C:/Program Files (x86)/Tesseract-OCR/tesseract.exe'   

        result = pytesseract.image_to_string(img)   
        res = result.split(':',1)
        res1 = res[1].split(':',1)
        res2 = res1[1].split(':',1)
        house_no = res2[1].partition('|')[0]

        name = res1[0].partition('\n')[0]
        
        dic['Name']=name
        dic['House No.']=house_no
        if res1[0].find("Husband's Name")!=-1:
            hn = res1[1].partition('\n')[0]
            dic["Husband's Name"]=hn
            dic["Father's Name"]=""
            dic["Mother's Name"]=""
        elif res1[0].find("Father's Name")!=-1:
            fn = res1[1].partition('\n')[0]
            dic["Father's Name"]=fn
            dic["Mother's Name"]=""
            dic["Husband's Name"]=""
        elif res1[0].find("Mother's Name")!=-1:
            mn = res1[1].partition('\n')[0]
            dic["Mother's Name"]=mn
            dic["Father's Name"]=""
            dic["Husband's Name"]=""
        ls.append(dic)
        

print(th)
print(ls)
```

```python
df = pd.DataFrame(ls)
```
