```python
# Generalized code for generating all PDFs of constituencies
# https://ceogoa.nic.in/PDF/EROLL/MOTHERROLL/2021/1/S05A1P1.pdf

base_url = "https://ceogoa.nic.in/PDF/EROLL/MOTHERROLL/2021/1/S05A"
extension = ".pdf"
# there are 40 assembly constituencies, making an array of the same 
assembly_url = []
for i in range(1,41):
  assembly_url.append(base_url+str(i)+"P")

#for all assembly constituencies there are different number of parts
number_of_parts = [40,48,43,40,42,44,40,35,39,41,30,43,39,37,40,48,50,56,50,44,45,45,42,30,46,30,41,40,43,42,40,40,39,45,45,49,47,48,46,59]

all_Goa_URLs = []

for i in range(0,40):
  for j in range(1,number_of_parts[i]+1):
    all_Goa_URLs.append(assembly_url[i]+str(j)+extension)

len(all_Goa_URLs)
```
