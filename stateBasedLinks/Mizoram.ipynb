```python
# Generalized code for generating all PDFs of constituencies
base_url = "https://ceo.mizoram.gov.in/ERollReportWithoutPhoto/S16/1-HACHHEK/S16A"
extension = ".pdf"
# there are 32 assembly constituencies, making an array of the same 
assembly_url = []
for i in range(1,41):
  assembly_url.append(base_url+str(i)+"P")

#for all assembly constituencies there are different number of parts
number_of_parts = [38, 40, 43, 26, 29, 34, 29, 30, 22, 29, 24, 20, 30, 19, 33, 26, 27, 22, 30, 29, 27, 28, 30, 25, 29, 29, 32, 27, 27, 28, 22, 21, 37, 35, 28, 70, 69, 42, 37, 44] 

all_Mizoram_URLs = []

for i in range(0,40):
  for j in range(1,number_of_parts[i]+1):
    all_Mizoram_URLs.append(assembly_url[i]+str(j)+extension)

len(all_Mizoram_URLs)

all_Mizoram_URLs

```
