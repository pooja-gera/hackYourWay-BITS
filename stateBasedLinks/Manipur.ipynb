```python
# Generalized code for generating all PDFs of constituencies

base_url = "https://ceomanipur.nic.in/eroll_manipur/ssr2023/Draft/1-Khundrakpam/MotherRoll/English/Integrated/Draft/S14A"
extension = ".pdf"
# there are 60 assembly constituencies, making an array of the same 
assembly_url = []

for i in range(1,61):
  assembly_url.append(base_url+str(i)+"P")

#for all assembly constituencies there are different number of parts
number_of_parts = [42,45,43,46,45,39,44,40,42,35,36,40,29,38,50,41,49,41,49,40,47,40,35,45,39,41,46,28,35,44,40,39,39,39,47,36,36,36,35,40,80,74,78,71,70,56,76,71,81,42,68,93,64,52,37,41,60,80,78,56]

all_Manipur_URLs = []

for i in range(0,40):
  for j in range(1,number_of_parts[i]+1):
    all_Manipur_URLs.append(assembly_url[i]+str(j)+extension)

len(all_Manipur_URLs)
```
