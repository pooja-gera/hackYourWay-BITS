```python
# Generalized code for generating all PDFs of constituencies
base_url = "https://ceosikkim.nic.in/ElectoralRolls/PDFView?PDFUrl=https://ceosikkim.nic.in/UploadedFiles/ElectoralRollPolling/S21A"
extension = ".pdf"
# there are 32 assembly constituencies, making an array of the same 
assembly_url = []
for i in range(1,33):
  assembly_url.append(base_url+str(i)+"P")

#for all assembly constituencies there are different number of parts
number_of_parts = [19,17,22,16,17,19,19,12,22,22,19,18,23,21,14,22,18,17,27,25,20,18,18,22,9,12,12,17,19,21,12,51]

all_Sikkim_URLs = []

for i in range(0,32):
  for j in range(1,number_of_parts[i]+1):
    all_Sikkim_URLs.append(assembly_url[i]+str(j)+extension)

len(all_Sikkim_URLs)
```
